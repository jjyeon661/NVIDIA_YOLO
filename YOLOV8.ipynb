```python
# 1. YOLOv8 라이브러리 설치
!pip install ultralytics
!pip install opencv-python pillow

# 2. 라이브러리 불러오기
import torch
from ultralytics import YOLO
import cv2
import matplotlib.pyplot as plt
from PIL import Image
from google.colab import files

# 3. 사용자로부터 이미지 업로드 받기
uploaded = files.upload()  # 바탕화면에서 이미지 업로드

# 4. 업로드된 파일 이름 가져오기
image_path = list(uploaded.keys())[0]  # 사용자가 업로드한 첫 번째 파일

# 5. YOLOv8 모델 로드 (사전 학습된 COCO 데이터셋 사용)
model = YOLO("yolov8n.pt")  # 'n'은 nano 모델로 가볍고 빠름

# 6. 사람 탐지 수행
results = model(image_path)

# 7. 결과 이미지 저장
output_path = "detected_people.jpg"
results[0].save(filename=output_path)

# 8. 결과 이미지 표시
img = Image.open(output_path)
plt.figure(figsize=(10, 10))
plt.imshow(img)
plt.axis("off")
plt.show()

```
