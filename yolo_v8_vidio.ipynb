```python
# 필요한 라이브러리 불러오기
import cv2
import torch
import os
import matplotlib.pyplot as plt
from ultralytics import YOLO
from IPython.display import display, Video
import yt_dlp

# 유튜브 영상 URL 입력
youtube_url = "https://youtube.com/shorts/8f4OnrAiRxA?si=iqD40s1HQcv2Pho1"  # 🎯 원하는 영상 링크로 변경하세요

# 저장할 파일 이름
video_filename = "youtube_video.mp4"

# yt-dlp를 이용한 YouTube 영상 다운로드
ydl_opts = {
    "format": "bestvideo+bestaudio/best",
    "outtmpl": video_filename,
    "quiet": True,
}
with yt_dlp.YoutubeDL(ydl_opts) as ydl:
    ydl.download([youtube_url])

print(f"📥 YouTube 영상 다운로드 완료: {video_filename}")

# YOLOv8 모델 로드 (yolov8n: 가벼운 모델, yolov8s/m/l/x 사용 가능)
model = YOLO("yolov8n.pt")

# 차량 관련 클래스 ID (YOLOv8에서 자동차, 버스, 트럭, 오토바이)
vehicle_classes = {2: "car", 3: "motorcycle", 5: "bus", 7: "truck"}

# YouTube에서 받은 영상으로 객체 감지 수행
results = model("youtube_video.mp4.webm", save=True, conf=0.5)  # 신뢰도 50% 이상만 감지

# YOLO가 저장한 결과 영상 찾기
output_dir = "runs/detect"
exp_folders = sorted(os.listdir(output_dir))
latest_exp = exp_folders[-1]  # 최신 실행 폴더
output_video_path = f"{output_dir}/{latest_exp}/{video_filename}"

print(f"🎬 처리된 영상 저장됨: {output_video_path}")

# Colab에서 결과 영상 재생
display(Video(output_video_path, embed=True))
```
